```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv("zomato.csv")
print(dataframe)
```

```python
dataframe
```

```python
import numpy as np

def handleRate(value):
    value = str(value).split('/')[0]   # e.g. "4.1/5" → "4.1"
    
    # agar value "NEW" ya "-" hai, to NaN return karo
    if value in ['NEW', '-']:
        return np.nan
    try:
        return float(value)
    except:
        return np.nan

# Apply function
dataframe['rate'] = dataframe['rate'].apply(handleRate)

print(dataframe['rate'].head())
```

```python
dataframe.info()
```

```python
print(dataframe.isnull().sum())
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel('Type of restaurant')
plt.ylabel('Votes')
```

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']

print('Restaurant(s) with the maximum votes:')
print(restaurant_with_max_votes)
```

```python
sns.countplot(x=dataframe['online_order'])
```

```python
# "rate" column me se '/5' hata ke numeric banao
dataframe['rate'] = dataframe['rate'].astype(str).str.replace('/5', '', regex=False)

# 'NEW' ya 'nan' jaise text ko NaN me convert karo
dataframe['rate'] = pd.to_numeric(dataframe['rate'], errors='coerce')

# Ab histogram plot karo
plt.hist(dataframe['rate'].dropna(), bins=5, color='skyblue', edgecolor='black')
plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```
